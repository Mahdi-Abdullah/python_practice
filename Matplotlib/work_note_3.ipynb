Scatter Plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

Scatter Plots with plt.plot

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black');
```

```python
plt.plot(x, y, '-ok');
```

```python
plt.plot(x, y, '-p', color='blue',
            linewidth=3, markersize=13,
            markerfacecolor='cyan',
            markeredgecolor='red',
            markeredgewidth=2)
plt.ylim(-1.2, 1.2);
```

Scatter Plots with plt.scatter

```python
plt.scatter(x, y, marker='p', c='red', s=50, edgecolors='blue');
```
